## Setup

```python
data_dir = "../data/"
output_dir = "../output/"
```

```python
import pandas as pd
df = pd.read_csv(data_dir + "taxonomy.csv", index_col = 0)
df.head()
```

```python
species = df.copy()
species[species < 0.00001] = 0
species.head()
```

```python
import pandas as pd
df = pd.read_csv(data_dir + "pathways.csv", index_col = 0)
pathways = df.copy()
pathways[pathways < 0.00001] = 0
pathways.head()
```

```python
full = pd.concat([pathways, species])
full
```

```python
with open(output_dir + "BothHealthy.txt") as f:
    BothHealthy = f.read().splitlines()
with open(output_dir + "BothUnhealthy.txt") as f:
    BothUnhealthy = f.read().splitlines()
BothHealthy
```

```python
healthyFeatures = full[full.index.isin(BothHealthy)]
unhealthyFeatures = full[full.index.isin(BothUnhealthy)]
display(healthyFeatures)
display(unhealthyFeatures)
```

```python
import numpy as np

def shannon(df):
    shannoned = np.log(df) * df * -1
    filled = shannoned.fillna(0)
    sums = list(filled.sum(axis=0, ))
    columns = list(df.columns)
    frame = pd.DataFrame(pd.Series(data=sums, index=columns)).T
    frame.index = ['shannon']
    return frame
```

```python
healthyShannon = shannon(healthyFeatures)
unhealthyShannon = shannon(unhealthyFeatures)
healthyShannon.iloc[:, :10]
```

```python
def richness(df):
    frame = pd.DataFrame((df > 0).sum()).T
    frame.index = ['Richness']
    return frame
```

```python
R_healthy = richness(healthyFeatures)
R_unhealthy = richness(unhealthyFeatures)
R_healthy.iloc[:, :10]
```

```python
healthySize = len(BothHealthy)
unhealthySize = len(BothUnhealthy)
healthySize, unhealthySize
```

```python
psi_H = pd.DataFrame((R_healthy / healthySize).values * healthyShannon.values, 
                      columns=healthyShannon.columns, index=["psi_H"])
psi_N = pd.DataFrame((R_unhealthy / unhealthySize).values * unhealthyShannon.values, 
                      columns=healthyShannon.columns, index=["psi_N"])
psi_N
```

```python
def GMHI(psi_H, psi_N):
    num = 0.0001 + psi_H.values
    den = 0.0001 + psi_N.values
    GMHI = pd.DataFrame(np.log10(num / den),
                        columns=psi_H.columns, index=["GMHI"]).T
    return GMHI
```

```python
test = GMHI(psi_H, psi_N)
test
```

```python
(test > 0).sum()
```

```python
isHealthy = pd.read_csv(data_dir + "isHealthy.csv", index_col=0)
isHealthy.T.head()
```

```python
def balanced_accuracy(GMHI, y):
    full = pd.concat([GMHI, y], axis=1)
    healthies = full[full['isHealthy']].iloc[:, [0]]
    predictedHealthy = int((healthies > 0).sum())
    unhealthies = full[~full['isHealthy']].iloc[:, [0]]
    predictedNot = int((unhealthies < 0).sum())
    P_H = predictedHealthy / len(healthies)
    P_N = predictedNot / len(unhealthies)
    print(P_H, P_N)
    chi = 0.5 * (P_H + P_N)
    return chi
```

```python
balanced_accuracy(test, isHealthy.T)
```
